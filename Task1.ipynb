```python
plt.hist(np.random.rand(100000), density=True, bins=1, histtype="step", color="blue", label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
