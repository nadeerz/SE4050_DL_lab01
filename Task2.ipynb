```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
dataframe.head()
```

```python
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
new_dataframe = dataframe[selected_columns]
```

```python
new_dataframe.head(20)
```

```python
mean_values = new_dataframe.mean()
std_dev_values = new_dataframe.std()
quartiles = new_dataframe.quantile([0.25, 0.50, 0.75])
print(mean_values)
print(std_dev_values)
print(quartiles)
```

```python
plt.figure(figsize=(20, 5))
for i, col in enumerate(selected_columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_dataframe[col], bins=10, rwidth=0.8)
    plt.title(col)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
data = new_dataframe['N_total']
plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.hist(data, bins=10, rwidth=0.8)
plt.title("N_total")
plt.xlabel("Value")
plt.ylabel("Frequency")
```
